## 예제 2-3 시리즈와 데이터프레임의 다양한 멤버 확인하기

```python
dir(pd.Series)
```

```python
s = set([ i for i in dir(pd.Series) if not i.startswith("_")])
```

```python
s
```

```python
d = set([ i for i in dir(pd.DataFrame) if not i.startswith("_")])
```

```python
s - d
```

```python
count = 0 
for i in (s - d ) :
    count += 1
    print(i, end=",  ") 
    if (count % 4 == 0) :
        print()
```

```python
count
```

```python
count = 0 
for i in (d - s ) :
    count += 1
    print(i, end=",  ") 
    if (count % 4 == 0) :
        print()
```

```python
count
```

```python
count = 0 
for i in (d & s ) :
    count += 1
    print(i, end=",  ") 
    if (count % 8 == 0) :
        print()
```

```python
count
```
